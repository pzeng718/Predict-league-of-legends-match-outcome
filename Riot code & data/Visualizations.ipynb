```python
import matplotlib.pyplot as plt
sns.scatterplot(data=players,x="wins",y="leaguePoints",hue='rank')
plt.xlim(0,800)
plt.ylim(0,1700)
```

```python
players['winrate'] = players['wins']/(players['wins']+players['losses'])
sns.scatterplot(data=players,x="winrate",y="leaguePoints",hue="rank")
plt.xlim(0.45,0.75)
plt.ylim(0,1700)
#players
```
