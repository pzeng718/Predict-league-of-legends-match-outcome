```python
df = df.drop(columns = ["web-scraper-order", "web-scraper-start-url",
                  "champ_page", "champ_page-href",
                  "counters_page", "counters_page-href",
                  "position pick rate", "ban rate"])
```

```python
for label in df.columns[2:]:
    df[label] = df[label].str.split()
```

```python
df.head(5)
```

```python
df.columns = ["champ_name", "counter_name", 
              "lane kill rate", "kill participation",
              "kda", "win ratio", 
              "position win rate","position pick rate", "ban rate"]
```

```python
df["champ_kda"] = df["kda"].str[0]
df["counter_kda"] = df["kda"].str[-3]
```

```python
for label in df.columns[2:-1]:
    if 'kda' in label:
        continue
    df["champ_"+label] = df[label].str[0]
    df["counter_"+label] = df[label].str[-1]
```

```python
df["championId"] = df["champ_name"].replace(championNamesToChampId)
df["counterId"] = df["counter_name"].replace(championNamesToChampId)
```

```python
for col in df[[
    "champ_lane kill rate", "counter_lane kill rate",
        'champ_kill participation', 'counter_kill participation',
       'champ_win ratio', 'counter_win ratio',
       'champ_position win rate', 'counter_position win rate',
       'champ_position pick rate', 'counter_position pick rate',
       'champ_ban rate', 'counter_ban rate']]:
    df[col] = df[col].str[:-1].astype("float") / 100.0
    df[col] = df[col].round(4)
```

```python
df = df[["champ_name", "counter_name",
    'championId','counterId', 
    'champ_kda', 'counter_kda',
   "champ_lane kill rate", "counter_lane kill rate",
   "champ_kill participation", "counter_kill participation",
   "champ_win ratio", "counter_win ratio",
   "champ_position win rate", "counter_position win rate",
   "champ_position pick rate", "counter_position pick rate",
   "champ_ban rate", "counter_ban rate"]]
```
