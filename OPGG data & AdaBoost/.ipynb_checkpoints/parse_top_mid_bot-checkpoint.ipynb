```python
df = df[["champ_name", "counter_name", "kda","lane kill rate", 
         "kill participation", "dmgtochamp", "first tower kill", "win ratio", 
         "position win rate", "position pick rate", "ban rate"]]
```

```python
for label in df.columns[2:]:
    df[label] = df[label].str.split()
```

```python
df["kda"].str[0]
```

```python
df["champ_kda"] = df["kda"].str[0]
df["counter_kda"] = df["kda"].str[-3]
```

```python
### after split, add new columns

for label in df.columns[3:-1]:
    if 'kda' in label:
        continue
    df["champ_"+label] = df[label].str[0]
    df["counter_"+label] = df[label].str[-1]
```

```python
df["championId"] = df["champ_name"].replace(championNamesToChampId)
df["counterId"] = df["counter_name"].replace(championNamesToChampId)
```

```python
pd.set_option('display.max_columns', None)
```

```python
# convert percentage (50.06%) to float (0.5006) 
# df["winratio"] = df["winratio"].str[:-1].astype("float") / 100.0 
for col in df[['champ_lane kill rate', 'counter_lane kill rate',
       'champ_kill participation', 'counter_kill participation',
       'champ_win ratio', 'counter_win ratio',
       'champ_position win rate', 'counter_position win rate',
       'champ_position pick rate', 'counter_position pick rate',
       'champ_ban rate', 'counter_ban rate']]:
    df[col] = df[col].str[:-1].astype("float") / 100.0
    df[col] = df[col].round(4)
```

```python
df = df[[ 'champ_name', 'counter_name',
    'championId','counterId', 
    'champ_kda', 'counter_kda', 
    'champ_lane kill rate', 'counter_lane kill rate',
    'champ_kill participation', 'counter_kill participation',
    'champ_dmgtochamp', 'counter_dmgtochamp', 
    'champ_first tower kill', 'counter_first tower kill', 
    'champ_win ratio', 'counter_win ratio',
    'champ_position win rate', 'counter_position win rate',
    'champ_position pick rate', 'counter_position pick rate',
    'champ_ban rate', 'counter_ban rate']]
```

```python
df['champ_kda'] = pd.to_numeric(df['champ_kda'])
df['counter_kda'] = pd.to_numeric(df['counter_kda'])
```

```python
df['champ_dmgtochamp'] = df['champ_dmgtochamp'].str.replace(',', '')
df['champ_dmgtochamp'] = pd.to_numeric(df['champ_dmgtochamp'])

df['counter_dmgtochamp'] = df['counter_dmgtochamp'].str.replace(',', '')
df['counter_dmgtochamp'] = pd.to_numeric(df['counter_dmgtochamp'])
```

```python
df['champ_first tower kill'] = df['champ_first tower kill'].str.replace("\'", "min")
df['champ_first tower kill'] = df['champ_first tower kill'].str.replace("\"", "s")
df['champ_first tower kill'] = pd.to_timedelta(df['champ_first tower kill'])

df['counter_first tower kill'] = df['counter_first tower kill'].str.replace("\'", "min")
df['counter_first tower kill'] = df['counter_first tower kill'].str.replace("\"", "s")
df['counter_first tower kill'] = pd.to_timedelta(df['counter_first tower kill'])
```
